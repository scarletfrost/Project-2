```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
confirmed_cases=pd.read_csv("us_confirmed.csv")

confirmed_cases.head()
```

```python
confirmed_cases = confirmed_cases.rename(columns={"Case":"Confirmed"})
confirmed_cases.head()
```

```python
us_cases = confirmed_cases.groupby(['Date', 'Province/State']).sum().reset_index()
us_cases
```

```python
us_cases.drop(['Long','Lat'], axis=1, inplace=True)
us_cases
```

```python
# us_cases_2 = us_cases.loc[us_cases['Confirmed'] != 0]
# us_cases_2
```

```python
deaths= pd.read_csv("us_deaths.csv")
deaths.head()
```

```python
deaths = deaths.rename(columns={"Case":"Deaths"})
deaths.head()
```

```python
us_deaths = deaths.groupby(['Date', 'Province/State']).sum().reset_index()
us_deaths
```

```python
us_deaths.drop(['Long','Lat'], axis=1, inplace=True)
us_deaths
```

```python
merged_file = pd.merge(us_deaths, us_cases, on=["Province/State","Date"])
merged_file
```

```python
covid_cases = merged_file.loc[merged_file['Confirmed'] != 0]
covid_cases
```
