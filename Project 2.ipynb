```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
confirmed_cases=pd.read_csv("us_confirmed.csv")

confirmed_cases.head()
```

```python
confirmed_cases.drop(['Long','Lat', 'Country/Region'], axis=1, inplace=True)
confirmed_cases.head()
```

```python
confirmed_cases = confirmed_cases.rename(columns={"Case":"Confirmed"})
confirmed_cases.head()
```

```python
confirmed_cases.dtypes
```

```python
confirmed_cases.groupby(['Date', 'Province/State'])
confirmed_cases.head()
```
